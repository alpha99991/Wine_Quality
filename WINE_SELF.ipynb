```python
sns.countplot(x='quality', data=wine)


#The above distribution shows the range for response variable (quality) is between 3 to 8.
#Let's create a new discreet, categorical response variable/feature ('rating') from existing 'quality' variable.
#i.e. bad: 1-4
#     average: 5-6
#     good: 7-10
```

```python
conditions = [
    (wine['quality'] >= 7),
    (wine['quality'] <= 4)
]
rating = ['good', 'bad']
wine['rating'] = np.select(conditions, rating, default='average')
wine.rating.value_counts()
```

```python
wine.groupby('rating').mean()
```

```python
y,X = dmatrices('quality ~ alcohol', data=wine, return_type='dataframe')
print("X:", type(X))
print(X.columns)
model=sm.OLS(y, X)
result=model.fit()
result.summary()
```

```python
model = sm.OLS.from_formula('quality ~ alcohol', data = wine)
results = model.fit()
print(results.params)
```

```python
wine['rate_code'] = (wine['quality'] > 4).astype(np.float32)
```

```python
y, X = dmatrices('rate_code ~ alcohol', data = wine)
sns.distplot(X[y[:,0] > 0, 1])
sns.distplot(X[y[:,0] == 0, 1])
```

```python
model = sm.Logit(y, X)
result = model.fit()
result.summary2()
```

```python
yhat = result.predict(X) > 0.955
print(sklearn.metrics.classification_report(y, yhat))
```

```python
# Classification using Sklearn's LogisticRegression:


model = sklearn.linear_model.LogisticRegression()
y,X = dmatrices('rate_code ~ alcohol + sulphates + citric_acid + fixed_acidity', data = wine)
model.fit(X, y)
yhat = model.predict(X)
print(sklearn.metrics.classification_report(y, yhat))
```

```python
"""
The accuracy matrix for sklearn's linear regression model for red wine quality prediction shows the overall
92% precision which is similar to previous statsmodel's average precision.
Also the precision for good wine (1) prediction is almost 96%.
But the precision is almost 0% for the bad type of wine (0) with sklearn's linear regression model.
Which is not a good sign for the analysis.

"""
```

```python
# Classification using Sklearn's RandomForestClassifier:

y, X = dmatrices('rate_code ~ alcohol', data = wine)
model = sklearn.ensemble.RandomForestClassifier()
model.fit(X, y)
yhat = model.predict(X)
print(sklearn.metrics.classification_report(y, yhat))
```

```python
"""
Here, with the accuracy matrix for sklearn's random forest classifier model for the prediction of red wine quality,
we can observe that the values have been improved significantly.
The precision for the prediction of bad quality wine (0) is almost 100% where as the precision for prediction of good quality
wine (1) is approximately 96%.
This sklearn's random forest classifier model also has the overall precision around 96%, which is far better
than the previous two models (i.e. statsmodel and sklearn's linear regression model)


"""
```
